```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats
```

```python
array1 = np.array([0, 38.5, 59, 97.4, 119.2, 129.5, 198.7, 248.7, 318, 438.5])
array2 = np.array([19.5, 15, 13.5, 23.3, 6.3, 2.5, 13, 1.8, 6.5, 1.8])
```

```python
# Корреляционный анализ
correlation_coefficient, p_value = stats.pearsonr(array1, array2)
# Выводим значения коэффициентов
print("Коэффициент корреляции:", correlation_coefficient)
print("Коэффициент детерминации:", correlation_coefficient ** 2)
print("Уровень значимости:", p_value)
```

```python
# Регрессионный анализ
model = sm.OLS(array2, array1) # создание модели
results = model.fit() # метод для сохранения значений анализа в переменной
print(results.summary()) # вывод результата
```

```python
# Создание графика корреляции
plt.scatter(array1, array2, label='Данные массивов')

# Создание линии регрессии
slope, intercept, _, _, _ = stats.linregress(array1, array2) # функция возвращает 5 значений, для построения линии регресси нужны 2
regression_line = intercept + slope * array1
plt.plot(array1, regression_line, color='red', label='Линия регрессии')

# Выводит коэффициент корреляции на график
plt.text(0.95, 0.75, f'Корреляция: {correlation_coefficient:.2f}', transform=plt.gca().transAxes, ha='right')

# Названия осей и графика
plt.xlabel('Эл. Активность')
plt.ylabel('Проницаемость сосудов')
plt.title('График корреляции вместе с линией регрессии')

# Легенда графика
plt.legend()
```

```python
# Уравнение регрессии
regression_equation = f"y = {slope:.2f}x + {intercept:.2f}"
print("Уравнение регрессии:")
print(regression_equation)
```
